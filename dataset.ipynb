In this notebook we produce the input of our model.  More specifically, we  produce triplets of "style, bodytype and occasion". The occasion and the bodytype are derived from the "neuralwork/fashion-style-instruct" dataset when the style list is generated by ChatGPT.

```python
from datasets import load_dataset

dataset = load_dataset("neuralwork/fashion-style-instruct")
```

```python
print(f"Train dataset size: {len(dataset['train'])}")
```

```python

# IMPORT THE PANDAS LIBRARY
# TO USE THE DATAFRAME TOOL
import pandas as pd
df = pd.DataFrame(data=dataset)
```

```python
df
```

```python
# Initialize empty lists for the new columns
indices = []
inputs = []
completions = []
contexts = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Extract data from the 'train' column
    train_data = row['train']
    # Append data to respective lists
    indices.append(index)  # Use the index generated by pandas
    inputs.append(train_data['input'])
    completions.append(train_data['completion'])
    contexts.append(train_data['context'])

# Create a new DataFrame with the extracted data
new_df = pd.DataFrame({
    'index': indices,
    'input': inputs,
    'completion': completions,
    'context': contexts
})

# Print the new DataFrame
print(new_df)
```

```python
# Assuming new_df is your DataFrame

# Set display options for pandas DataFrame
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000)        # Set maximum display width
pd.set_option('display.max_rows', None)     # Show all rows

# Print the DataFrame
print(new_df)
```

```python
# Assuming new_df is your DataFrame

# Extract the desired part of the text from the 'context' column
new_df['extracted_context'] = new_df['context'].str.split("I'm going to ").str[1].str.split(" as").str[0]

# Print the DataFrame with the extracted context
print(new_df)
```

```python
# Assuming new_df is your DataFrame

# Extract the desired part of the text from the 'context' column
new_df['event'] = new_df['extracted_context'].str.split(".", n=1).str[0]

# Print the DataFrame with the extracted context
print(new_df)
```

```python
# Slice the DataFrame to the first 2740 rows
new_df_sliced = new_df.iloc[:2741]

# Print the sliced DataFrame
print(new_df_sliced)
```

```python
new_df_sliced['extracted_type'] = new_df_sliced['input'].str.split("I'm ").str[1].str.split(" who").str[0]

new_df_sliced['type'] = new_df_sliced['extracted_type'].str.split(".", n=1).str[0]


# Print the DataFrame with the extracted context
print(new_df_sliced)
```

```python
import csv

def generate_touples(styles_list, simple_types_list):
    touples = []
    for style in styles_list:
        for typ in simple_types_list:
                touples.append((style, typ))
    return touples

def write_to_csv(triplets, filename):
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Style', 'Type', 'Event'])
        writer.writerows(triplets)

def generate_triples(touples, unique_events_list):
    triples = []
    repeated_tuples = touples * 10
    for tup in repeated_tuples:
      event = random.choice(unique_events_list)
      triples.append(tup + (event,))
    return triples

# Generate triplets
touples = generate_touples(styles_list, simple_types_list)
triples = generate_triples(touples, unique_events_list)

complex_types_touples = generate_touples(styles_list, small_complex_types_list)
complex_types_triples = generate_triples(complex_types_touples, unique_events_list)
# Write to CSV
write_to_csv(triples, 'simple_types.csv')
write_to_csv(complex_types_triples, 'complex_types.csv')
```
